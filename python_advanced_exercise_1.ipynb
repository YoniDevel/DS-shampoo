    1. Write a Date class
        a. you can save only milliseconds as data member
        b. these operators must work: < > == != + - ...
        c. you can access fields using a key: date['day'], date['month']

```python
def is_leap_year(year: int) -> bool:
  if year % 4 == 0:
    if year % 100 == 0:
      return year % 400 == 0
    return True
  return False

def handle_month(month: int, year: int) -> int:
  day_in_mili = 1000 * 60 * 60 * 24
  return day_in_mili * (
      31 if month in [1,3,5,7,8,10,12]
      else 30 if month in [4,6,9,11]
      else 29 if is_leap_year(year)
      else 28
    )

class Date:

  def __init__(self, year=1, month=1, day=1, hour=0, minute=0, second=0, time=0):
    if time != 0: self.time = time
    else:
      year_in_mili = 1000 * 60 * 60 * 24 * 365
      month_in_mili = handle_month(month, year)
      day_in_mili = 1000 * 60 * 60 * 24
      hour_in_mili = 1000 * 60 * 60
      minute_in_mili = 1000 * 60
      second_in_mili = 1000

      self.time = (year*year_in_mili + month*month_in_mili + day*day_in_mili
      + hour*hour_in_mili + minute*minute_in_mili + second*second_in_mili)

  def __eq__(self, other):
      return self.time == other.time

  def __ne__(self, other):
    return self.time != other.time

  def __lt__(self, other):
      return self.time < other.time

  def __gt__(self, other):
      return self.time > other.time

  def __add__(self, other):
      return Date(time=self.time + other.time)

  def __sub__(self, other):
      return Date(time=self.time - other.time)

  def __getitem__(self, key):
    year_in_mili = 1000 * 60 * 60 * 24 * 365
    month_in_mili = 1000 * 60 * 60 * 24 * 31
    day_in_mili = 1000 * 60 * 60 * 24
    hour_in_mili = 1000 * 60 * 60
    minute_in_mili = 1000 * 60
    second_in_mili = 1000

    if key == 'year':
      return math.floor(self.time/year_in_mili)
    elif key == 'month':
      months = math.floor(self.time%year_in_mili/month_in_mili)
      return months if months > 0 else 12
    elif key == 'day':
      days = math.floor(self.time%year_in_mili%month_in_mili/day_in_mili)
      return days if days > 0 else 30
    elif key == 'hour':
      return math.floor(self.time%year_in_mili%month_in_mili%day_in_mili/
                        hour_in_mili)
    elif key == 'minute':
      return math.floor(self.time%year_in_mili%month_in_mili%day_in_mili%
                        hour_in_mili/minute_in_mili)
    elif key == 'second':
      return math.floor(self.time%year_in_mili%month_in_mili%day_in_mili%
                        hour_in_mili%minute_in_mili/second_in_mili)
    else: return self.time

  def __str__(self):
    returned_string = ""
    for key in ['year', 'month', 'day']:
      returned_string += f'{self[key]}-'
    returned_string = returned_string[:-1] + ' '
    for key in ['hour', 'minute', 'second']:
      returned_string += f'{self[key]}:'
    returned_string = returned_string[:-1]
    return returned_string


d1 = Date(2004, 1, 3)
print(d1.time)
print(d1['year'])
print(d1['month'])
print(d1['day'])
print(d1.__str__())
```
